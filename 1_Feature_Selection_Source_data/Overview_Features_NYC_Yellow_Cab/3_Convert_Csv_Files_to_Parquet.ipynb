### NYC Yellow Cab 2009

#### Convert all files to parquet Files

```python
dtypes={"VendorID":object,"Passenger_Count":np.uint8,"Trip_Distance":np.float32,"Pickup_Lon":np.float32,"Pickup_Lat":np.float32,"RateCodeID":object,"Store_and_Forward":object,"Dropoff_Lon":np.float32,"Dropoff_Lat":np.float32,"Payment_Type":object,"Fare_Amt":np.float16,"Surcharge":np.float16,"Mta_Tax":np.float16,"Tip_Amt":np.float16,"Tolls_Amt":np.float16,"Total_Amt":np.float16}
# Rename the Columns for uniform Database
names=["VendorID","Pickup_DateTime","Dropoff_DateTime","Passenger_Count","Trip_Distance","Pickup_Lon","Pickup_Lat","RateCodeID","Store_and_Forward","Dropoff_Lon","Dropoff_Lat","Payment_Type","Fare_Amt","Surcharge","Mta_Tax","Tip_Amt","Tolls_Amt","Total_Amt"]
# Set Path for Parquet Files
path="F:/1_Bachelorarbeit/1_Daten/1_Yellow_Cab_Trips_2009-June_2016/Parquet_Files/2009/"
for csv_file in glob.glob(path_2009):
    df = dd.read_csv(csv_file,dtype=dtypes,names=names,skiprows=1,parse_dates=["Pickup_DateTime","Dropoff_DateTime"])
    df.to_parquet(os.path.join(path, csv_file[83:90] + "." + "parquet"),engine="fastparquet")
```

```python
# Testing the Parquet Files
df=pd.read_parquet("F:/1_Bachelorarbeit/1_Daten/1_Yellow_Cab_Trips_2009-June_2016/Parquet_Files/2009/2009-01.parquet",engine="fastparquet")
df.head()
```
