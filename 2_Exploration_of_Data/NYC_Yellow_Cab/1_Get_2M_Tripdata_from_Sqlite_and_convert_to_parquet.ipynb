```python
import sqlite3
import pandas as pd
import numpy as np
import fastparquet
import os
```

### Connect to Database and create Cursor 

```python
path="F:/1_Bachelorarbeit/1_Daten/1_Yellow_Cab_Trips_2009-June_2016/4_Sqlite/NYC_Yellow_Cab_Tripdata"
# Create Database or if Name is there open Database
connection = sqlite3.connect(path)
# Create Cursor for SQL-Commands
cursor=connection.cursor()
```

### Sample 2.000.000 Random Entries of all Tripdata Tables

```python
dtypes={"Pickup_DateTime":np.datetime64,"Dropoff_DateTime":np.datetime64,"Passenger_Count":np.uint8,"Trip_Distance":np.float16,"Pickup_Lon":np.float32,"Pickup_Lat":np.float32,"RateCodeID":np.uint8,"Dropoff_Lon":np.float32,"Dropoff_Lat":np.float32,"Fare_Amt":np.float16}
# Give me 250000 random sample from table tripdata_2009
sql = "SELECT * FROM tripdata_2009 ORDER BY RANDOM() LIMIT 250000;"  
df_origin = pd.read_sql_query(sql, connection, index_col="Trip_ID")
# Replace all None Values of RateCodeID with Zero -> Because 0 is for none RateCodeID
df_origin["RateCodeID"].fillna(value=0, inplace=True)
```
