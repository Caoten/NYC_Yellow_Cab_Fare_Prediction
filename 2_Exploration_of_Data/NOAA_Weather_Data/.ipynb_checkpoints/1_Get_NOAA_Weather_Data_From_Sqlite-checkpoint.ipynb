```python
import sqlite3
import pandas as pd
import numpy as np
import fastparquet
import os
```

### Connect to Database and create Cursor 

```python
path="F:/1_Bachelorarbeit/1_Daten/1_Yellow_Cab_Trips_2009-June_2016/4_Sqlite/NYC_Yellow_Cab_Tripdata"
# Create Database or if Name is there open Database
connection = sqlite3.connect(path)
# Create Cursor for SQL-Commands
cursor=connection.cursor()
```

### Get all NOAA Weather Data Entries of Table weather_data

```python
dtypes={"DATE":np.datetime64,"AWND":np.float16,"PRCP":np.float16,"SNOW":np.float16,"SNWD":np.float16,"TAVG":np.float16,"TMAX":np.float16,"TMIN":np.float16}
# Give me all data from table weather data
sql = "SELECT * FROM weather_data;"  
df_origin = pd.read_sql_query(sql, connection, index_col="Weather_ID")
```

### Replace NaN's with 0 in WT01-WT22

```python
# Replace all NaN with mean Value and set is_dummy to True
df_origin[["WT01","WT02","WT03","WT04","WT05","WT06","WT07","WT08","WT09","WT11","WT13","WT14","WT16","WT17","WT18","WT19","WT22"]]=df_origin[["WT01","WT02","WT03","WT04","WT05","WT06","WT07","WT08","WT09","WT11","WT13","WT14","WT16","WT17","WT18","WT19","WT22"]].fillna(value=0)
```

```python
df_origin
```

## Change Values of WT01-WT22 to Boolean True,False

```python
# Replace all NaN with mean Value and set is_dummy to True
df_origin[["WT01","WT02","WT03","WT04","WT05","WT06","WT07","WT08","WT09","WT11","WT13","WT14","WT16","WT17","WT18","WT19","WT22"]]=df_origin[["WT01","WT02","WT03","WT04","WT05","WT06","WT07","WT08","WT09","WT11","WT13","WT14","WT16","WT17","WT18","WT19","WT22"]].replace({0:False,1:True})
```
