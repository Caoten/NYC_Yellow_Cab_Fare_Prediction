### Connect to Database and create Cursor 

```python
path="F:/1_Bachelorarbeit/1_Daten/1_Yellow_Cab_Trips_2009-June_2016/4_Sqlite/NYC_Yellow_Cab_Tripdata"
# Create Database or if Name is there open Database
connection = sqlite3.connect(path)
# Create Cursor for SQL-Commands
cursor=connection.cursor()
```

### Get all NOAA Weather Data Entries of Table weather_data

```python
dtypes={"DATE":np.datetime64,"TAVG":np.float16,"TMAX":np.float16,"TMIN":np.float16}
# Give me all data from table weather data
sql = "SELECT * FROM weather_data;"  
df_origin = pd.read_sql_query(sql, connection, index_col="Weather_ID")
```

#### Change NAN Values of Weathertype to 0

```python
df_origin["WT01"].fillna(0,inplace=True)
df_origin["WT11"].fillna(0,inplace=True)
df_origin["WT16"].fillna(0,inplace=True)
df_origin["WT18"].fillna(0,inplace=True)
```

```python
df_origin["WT01"]=df_origin["WT01"].astype("uint8")
df_origin["WT11"]=df_origin["WT11"].astype("uint8")
df_origin["WT16"]=df_origin["WT16"].astype("uint8")
df_origin["WT18"]=df_origin["WT18"].astype("uint8")
```
