### Read Parquet Files of NYC Yellow Cab 2009 and load in Database Table Tripdata

```python
# Columns to read from parquet File
columns=["Pickup_DateTime","Dropoff_DateTime","Passenger_Count","Trip_Distance","Pickup_Lon","Pickup_Lat","RateCodeID","Dropoff_Lon","Dropoff_Lat","Fare_Amt"]
```

```python
# Load all Parquet File with selected Columns in Dataframe and load to sqlite
for csv_file in glob.glob(path_2009):
    df=pd.read_parquet(csv_file,engine="fastparquet",columns=columns)
    df.to_sql(name="tripdata_2009", con=connection, if_exists="append", index=False) 
```

```python
# Look into Table to confirm that the data are loaded
sql = "SELECT * FROM tripdata_2009 LIMIT 10;"  
df = pd.read_sql_query(sql, connection)
df.head(10)
```

```python
sql="SELECT COUNT(*) FROM tripdata_2009" 
pd.read_sql_query(sql, connection)
```
