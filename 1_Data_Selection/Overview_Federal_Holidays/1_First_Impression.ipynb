```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
path="F:/1_Bachelorarbeit/1_Daten/11_Federal-Holidays-US/usholidays.csv"
dtypes={"Holiday":object}
df_us_holidays = pd.read_csv(path,dtype=dtypes,parse_dates=["Date"],doublequote=True)
```

```python
df_us_holidays
```

#### Observation:
#### 1. There are Dates outside of 2009-2016
#### 2. The Dates are unsorted
#### 3. Column Unnamed: 0 is unnecessary 

```python
df_us_holidays.info()
```

```python
# First Step is to sort the Dataframe on Date
df_us_holidays=df_us_holidays.sort_values(by="Date")
```

```python
df_us_holidays
```

```python
# Second Step is to exclude Datetime rows which are not between 2009 and June 2016
mask = (df_us_holidays['Date'] >= '2009-01-01') & (df_us_holidays['Date'] <= '2016-06-30')
df_us_holidays=df_us_holidays.loc[mask]
```

```python
df_us_holidays
```

```python
# Third Step is to remove column Unnamed: 0
df_us_holidays.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df_us_holidays
```

```python
df_us_holidays.to_csv("F:/1_Bachelorarbeit/1_Daten/1_Yellow_Cab_Trips_2009-June_2016/3_US_Federal_Holidays/usholidays_new.csv",index=False)
```
