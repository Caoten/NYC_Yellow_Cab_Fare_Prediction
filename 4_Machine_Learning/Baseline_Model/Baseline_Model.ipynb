```python
# Create Training and Testset for the 3 Datasets
X_train_without, X_test_without, y_train_without, y_test_without = train_test_split(data_without.drop(columns=["Fare_Rate"]), data_without["Fare_Rate"], test_size=0.2, random_state=1)
X_train_with, X_test_with, y_train_with, y_test_with = train_test_split(data_with.drop(columns=["Fare_Rate"]), data_with["Fare_Rate"], test_size=0.2, random_state=1)
X_train_with_external, X_test_with_external, y_train_with_external, y_test_with_external = train_test_split(data_with_external.drop(columns=["Fare_Rate"]), data_with_external["Fare_Rate"], test_size=0.2, random_state=1)
```

```python
# Resampling Training Datasets
X_train_without_undersample,y_train_without_undersample=Resampling.random_undersample(X_train_without,y_train_without)
X_train_without_oversample,y_train_without_oversample=Resampling.random_oversample(X_train_without,y_train_without)

X_train_with_undersample,y_train_with_undesample=Resampling.random_undersample(X_train_with,y_train_with)
X_train_with_oversample,y_train_with_ovesample=Resampling.random_oversample(X_train_with,y_train_with)

X_train_with_external_undersample,y_train_with_external_external_undersample=Resampling.random_undersample(X_train_with_external,y_train_with_external)
X_train_with_external_external_oversample,y_train_with_external_oversample=Resampling.random_oversample(X_train_with_external,y_train_with_external)
```

### Baseline Model

```python
# Get Index of the most frequent Fare_Bin
Most_Frequent_Fare_Rate_Train=y_train_without.value_counts().idxmax()
Most_Frequent_Fare_Rate_Test=y_test_without.value_counts().idxmax()
# Create Numpy Array of Prediction
y_train_without_pred=np.empty(shape=len(y_train_without))
y_train_without_pred.fill(Most_Frequent_Fare_Rate_Train)
y_test_without_pred=np.empty(shape=len(y_test_without))
y_test_without_pred.fill(Most_Frequent_Fare_Rate_Test)
```

### Evaluation for Training Data

```python
print("Accuracy: {:.3f} %".format(accuracy_score(y_train_without,y_train_without_pred)*100))
```

```python
# Plot Heatmap
plt.figure(figsize=(12,12))
corr = sns.heatmap(confusion_matrix(y_train_without,y_train_without_pred), annot = True)
plt.title('Confusion Matrix',fontsize=20)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
```
